```python
import pandas as pd
```

```python
from alpha_vantage.timeseries import TimeSeries
ts = TimeSeries(key='0DKKRV0PAG53Z1Q1')
# # Get json object with the intraday data and another with  the call's metadata
# data, meta_data = ts.get_intraday('GOOGL')
```

```python
import numpy as np
```

```python
class ScriptData:
    def __init__(self):
        self.stocks_dfs = {}
    
    def __setitem__(self, key, value):
        self.stocks_dfs[key] = value
    
    def __getitem__(self, key):
        return self.stocks_dfs[key]
    
    def __contains__(self,param):
          return True if param in self.stocks_dfs.keys() else False
    
    def fetch_intraday_data(self, stock):
        data, _ = ts.get_intraday(stock)
        return data
    def convert_intraday_data(self, stock):
        data = ScriptData.fetch_intraday_data(self, stock)
        time_stamp = []
        stock_info = []
        d = list(data.items())

        for i in range(len(d)):
            time_stamp.append((d[i][0]))
        time_stamp = {"timestamp":time_stamp}
        df1 = pd.DataFrame.from_dict(time_stamp)
        for i in range(len(d)):
            stock_info.append((d[i][1]))
        df2 = pd.DataFrame.from_dict(stock_info)
        df = pd.concat([df1,df2], axis = 1)
        df = df.rename(columns={"1. open": "open", "2. high": "high", "3. low": "low", "4. close": "close", "5. volume": "volume"})
        ScriptData.__setitem__(self, stock, df)
#         self.stocks_dfs[stock] = df
        return df

            
```

```python
script_data = ScriptData()
```

```python
def indicator1(df, timeperiod):
    timestamp = df['timestamp'].to_frame()
    indicator = df['close'].to_frame()
    indicator = indicator.rolling(timeperiod).mean()
    indicator_df = pd.concat([timestamp, indicator], axis = 1)
    indicator_df = indicator_df.rename(columns={'close':'indicator'})
    return indicator_df
    
    
```

```python
indicator1(script_data['GOOGL'], 5)
```

```python
class Strategy(ScriptData):
    def __init__(self, stock):
        super().__init__()
        self.stk = stock
    
    def get_script_data(self):
        close = Strategy.convert_intraday_data(self, self.stk)[["timestamp", "close"]]
        indicator = indicator1(Strategy.convert_intraday_data(self, self.stk), 5)["indicator"].to_frame()
        close["close"] = close["close"].astype(float)
        return close, indicator
    
    def get_signals(self):
        df = pd.concat([Strategy.get_script_data(self)[0], Strategy.get_script_data(self)[1]], axis = 1)
        df.dropna(inplace=True)
        df["signal"] = pd.DataFrame(np.where(df['indicator']>df['close'], 1, 0)).diff()
        df["signal"] = df["signal"].apply(lambda x: "BUY" if x == 1 else("SELL" if x == -1 else np.NaN))
        df.dropna(inplace=True)
        return df
```
